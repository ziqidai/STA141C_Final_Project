### Lu Decomposition

```python
t = time.time()
t = time.time()
beta_hat = np.linalg.solve(X.T.dot(X) ,X.T.dot(Y))   # beta hat given
print(np.round_(time.time()-t, 3),'sec elapsed')       # counts time
print(beta_hat)
```

---

### QR Decomposition

```python
t = time.time() 
new_matrix = np.c_[X,Y] 
Qq , Rr = slg.qr(new_matrix, mode='economic')   
R = Rr[:-1,:-1]         # class_note
r = Rr[:-1,-1]
beta_hat2 = slg.solve(R,r)   # beta_hat
print(np.round_(time.time()-t, 3),'sec elapsed')  
pd.DataFrame(beta_hat2)
```

```python
def svd(X):
  n, m = X.shape
  U, Sigma, Vh = np.linalg.svd(X, 
      full_matrices=False,
      compute_uv=True)
  X_svd = np.dot(U, np.diag(Sigma))
  return X_svd
```
